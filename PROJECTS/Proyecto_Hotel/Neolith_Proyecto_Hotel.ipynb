### Desafío: Optimización de Precios para Proyecto de Hotel

**Contexto:**
Neolith está considerando una nueva oportunidad de proyecto que implica el suministro de materiales de piedra sinterizada de alta calidad para un prestigioso proyecto de hotel. Este proyecto abarca varias aplicaciones de nuestros productos, incluidas soluciones para suelos, encimeras/mesas y fachadas ventiladas.

**Productos y Costes:**
- Fusion: 22€/m2
- Timber: 22€/m2
- Colorfeel: 28€/m2
- Iron: 21€/m2

**Cantidad Producto Seleccionado Hotel**
- Fusion: 5.000m2
- Timber: 2.000m2
- Colorfeel: 300m2
- Iron: 1.500m2

**Requisitos del Proyecto:**
- Los productos Fusion y Timber se utilizarán para el suelo.
- Iron se seleccionará para encimeras/mesas.
- Colorfeel se aplicará en la fachada ventilada.
- Se ha determinado que la oferta para el suelo se realizará a un precio por debajo del coste, específicamente a 20€/m2, debido a condiciones competitivas del mercado.

**Objetivo:**
A pesar de la reducción de precios en ciertas áreas, el equipo ejecutivo de Neolith aspira a alcanzar un margen mínimo del 14% en el proyecto global. Este margen es crucial para mantener la sostenibilidad financiera y apoyar la estrategia de crecimiento de Neolith.

**Tu Reto:**
Como Pricing Analyst, tu misión es desarrollar una estrategia de precios detallada que permita a Neolith alcanzar el objetivo de margen mínimo del 14% a través de:
- Análisis detallado del coste de cada producto.
- Estrategias de precios para cada aplicación de producto, teniendo en cuenta el requisito de vender a pérdida en el segmento del suelo.
- Optimización de precios para productos utilizados en encimeras/mesas y fachadas, que compense las pérdidas incurridas en el suelo y contribuya al margen global deseado.
- Presentación de un rango de precios negociables que maximice la rentabilidad sin comprometer la competitividad en el mercado.

**Entregables:**
- Un análisis detallado del coste y beneficio por producto y total del proyecto.
- Recomendaciones de precios para los productos Iron y Colorfeel que aseguren el cumplimiento del margen objetivo.
- Justificación de tus estrategias basada en análisis de costes y un entendimiento profundo de la dinámica del mercado.

# Importación de Paquetes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Datos

```python
productos = ['Fusion', 'Timber', 'Iron', 'Colorfeel']
costes = np.array([22, 22, 21, 28])
precios_venta_solicitados = np.array([20, 20, 45, 55]) 
m2 = np.array([5000, 2000, 300, 1500]) 
```

# Calculo de Costes

```python
# Calculo de costes y ventas totales
ventas_esperadas = precios_venta_solicitados * m2
coste_total = np.sum(costes * m2)
venta_total = np.sum(ventas_esperadas)
margen = (venta_total - coste_total) / venta_total
```

# DataFrame y Visualización

```python
# DataFrame para visualización
df = pd.DataFrame({
    'Producto': productos + ['Total'],
    'Coste (€)': list(costes * m2) + [coste_total],
    'Venta Esperada (€)': list(ventas_esperadas) + [venta_total],
    'M2': list(m2) + [np.sum(m2)]
})
df['Beneficio (€)'] = df['Venta Esperada (€)'] - df['Coste (€)']
df
```

```python
# Calculamos el margen de toda la operación usando el total de costes y ventas esperadas
margen_operacion = (df.loc[df['Producto'] == 'Total', 'Venta Esperada (€)'].values[0] - 
                    df.loc[df['Producto'] == 'Total', 'Coste (€)'].values[0]) / df.loc[df['Producto'] == 'Total', 'Venta Esperada (€)'].values[0]

# Convertimos el margen a porcentaje
margen_operacion_porcentaje = margen_operacion * 100

margen_operacion_porcentaje

```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 12))

# Gráfica de barras para los costes y ventas por producto sin grid
ax.bar(df['Producto'], df['Coste (€)'], color='red', label='Coste')
ax.bar(df['Producto'], df['Venta Esperada (€)'], color='green', alpha=0.5, label='Venta Esperada')
ax.set_title('Coste vs Venta Esperada por Producto')
ax.set_ylabel('Euros (€)')
ax.legend()
ax.grid(False)  

```
