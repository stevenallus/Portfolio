# Análisis Costes Producción

**Datos Iniciales**

```python
precio_venta = 20 
margen_ganancia_deseado = 5  
coste_actual_por_m2 = 22  
produccion_actual_m2 = 2000
```

Calculamos el nuevo coste de producción por m2 para alcanzar el margen de ganancia deseado

```python
coste_nuevo = precio_venta - (margen_ganancia_deseado / 100) * precio_venta
```

Dado que el proceso productivo es el 50% del coste, cualquier reducción en el coste de producción afecta solo a esta parte

Calculamos la reducción necesaria en el coste del proceso productivo para alcanzar el costo nuevo

```python
reduccion_coste_necesaria = coste_actual_por_m2 - coste_nuevo
```

Calculamos el porcentaje de reducción en el coste del proceso productivo sobre el 50% del coste original

```python
porcentaje_reduccion_coste = (reduccion_coste_necesaria / coste_actual_por_m2) * 100
```

Cada aumento del 10% en producción da una mejora del 5% en el coste, calculamos el aumento necesario en producción

Primero encontramos la mejora necesaria en porcentaje del costo para igualar a la reducción necesaria

```python
mejora_coste_por_aumento_produccion = porcentaje_reduccion_coste / 5
```

Calculamos el aumento necesario en la producción

```python
aumento_produccion_necesario = mejora_coste_por_aumento_produccion * 10
```

```python
print('El aumento productivo es: {:.2f}%'.format(aumento_produccion_necesario)),
print('El nuevo coste con el aumento productivo es: ',coste_nuevo,'€/m2')
```

Calculamos la producción adicional necesaria

```python
produccion_adicional = produccion_actual_m2 * (aumento_produccion_necesario / 100)

nueva_produccion_total_diaria = produccion_actual_m2 + produccion_adicional

produccion_adicional, nueva_produccion_total_diaria
```

```python
print('La producción adicional es de: {:.0f} m2'.format(produccion_adicional)),
print('La nueva producción total diria es de: {:.0f} m2'.format(nueva_produccion_total_diaria))
```

## Punto de Equilibrio

```python
def calcular_punto_equilibrio(coste_fijo_total, coste_variable_unitario, precio_venta):
    # Punto de equilibrio en unidades
    unidades_equilibrio = coste_fijo_total / (precio_venta - coste_variable_unitario)
    
    # Punto de equilibrio en ingresos
    ingresos_equilibrio = unidades_equilibrio * precio_venta
    
    return unidades_equilibrio, ingresos_equilibrio

# Ejemplo de cálculo para un precio de venta específico
precio_venta_ejemplo = 20  # Precio de venta por unidad
unidades_equilibrio, ingresos_equilibrio = calcular_punto_equilibrio(coste_fijo_total, coste_variable_unitario, precio_venta_ejemplo)

print(f"Punto de equilibrio en unidades: {unidades_equilibrio:.2f}")
print(f"Punto de equilibrio en ingresos: {ingresos_equilibrio:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir un rango de costos fijos y costos variables unitarios para el análisis de sensibilidad
costos_fijos = np.linspace(10000, 50000, 5) 
costos_variables_unitarios = np.linspace(4, 6, 5) 
precio_venta = 20  

# Matrices para almacenar los resultados
unidades_equilibrio_sensibilidad = np.zeros((len(costos_fijos), len(costos_variables_unitarios)))
ingresos_equilibrio_sensibilidad = np.zeros_like(unidades_equilibrio_sensibilidad)

# Calcular el punto de equilibrio para cada combinación de costos fijos y variables
for i, cf in enumerate(costos_fijos):
    for j, cv in enumerate(costos_variables_unitarios):
        unidades, ingresos = calcular_punto_equilibrio(cf, cv, precio_venta)
        unidades_equilibrio_sensibilidad[i, j] = unidades
        ingresos_equilibrio_sensibilidad[i, j] = ingresos

# Graficar los resultados
fig, ax = plt.subplots(1, 2, figsize=(18, 6))


cf, cv = np.meshgrid(costos_variables_unitarios, costos_fijos)
contour = ax[0].contourf(cv, cf, unidades_equilibrio_sensibilidad, cmap="viridis")
fig.colorbar(contour, ax=ax[0])
ax[0].set_title('Unidades de Equilibrio vs. Costes Fijos y Variables')
ax[0].set_xlabel('Coste Variable Unitario')
ax[0].set_ylabel('Coste Fijo Total')


contour = ax[1].contourf(cv, cf, ingresos_equilibrio_sensibilidad, cmap="magma")
fig.colorbar(contour, ax=ax[1])
ax[1].set_title('Ingresos de Equilibrio vs. Costos Fijos y Variables')
ax[1].set_xlabel('Coste Variable Unitario')
ax[1].set_ylabel('Coste Fijo Total')

plt.tight_layout()
plt.show()
```

- **Unidades de Equilibrio vs. Costes Fijos y Variables** (Gráfico Izquierdo): Muestra cómo el número de unidades necesarias para alcanzar el punto de equilibrio aumenta con costes fijos y variables más altos. La variación en el color indica diferencias en la cantidad de unidades necesarias, con áreas más claras representando más unidades requeridas.

- **Ingresos de Equilibrio vs. Costes Fijos y Variables** (Gráfico Derecho): Presenta cómo el total de ingresos necesarios para alcanzar el punto de equilibrio se ve afectado por los mismos factores. Áreas más claras indican mayores ingresos requeridos para cubrir los costes y alcanzar la rentabilidad.

**Conclusiones Clave**:

- **Sensibilidad a los Costes**: Estos gráficos subrayan la sensibilidad del punto de equilibrio a variaciones en los costes fijos y variables. A medida que cualquiera de estos costes aumenta, también lo hace la dificultad para alcanzar el punto de equilibrio, ya sea en términos de unidades vendidas o ingresos necesarios.

## Simulaciones de Escenarios

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def simular_escenarios(coste_fijo_total, coste_variable_unitario, rangos_produccion, rangos_precio_venta):
    resultados = []

    for produccion in rangos_produccion:
        for precio_venta in rangos_precio_venta:
            ingresos = produccion * precio_venta
            coste_total = coste_fijo_total + (coste_variable_unitario * produccion)
            margen_ganancia = ingresos - coste_total
            resultados.append({
                "Producción": produccion,
                "Precio de Venta": precio_venta,
                "Ingresos": ingresos,
                "Coste Total": coste_total,
                "Margen de Ganancia": margen_ganancia
            })

    return pd.DataFrame(resultados)

# Ejemplo de uso
coste_fijo_total = 10000  
coste_variable_unitario = 5 

# Definir los rangos de producción y precios de venta a simular
rangos_produccion = np.arange(1000, 5001, 1000)  # De 1000 a 5000 unidades, en pasos de 1000
rangos_precio_venta = np.arange(10, 21, 2)  # De 10 a 20, en pasos de 2

# Realizar la simulación
resultados_simulacion = simular_escenarios(coste_fijo_total, coste_variable_unitario, rangos_produccion, rangos_precio_venta)

# Mostrar los resultados
resultados_simulacion
```

```python
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


sns.set(style="dark", rc={"axes.facecolor": "black", "grid.color": "none"})
plt.style.use("dark_background")

fig, ax = plt.subplots(figsize=(12, 8))

palette = sns.color_palette("Blues_r", n_colors=len(rangos_produccion))

for (produccion, grupo), color in zip(resultados_simulacion.groupby('Producción'), palette):
    sns.lineplot(x='Precio de Venta', y='Margen de Ganancia', data=grupo, label=f'Producción: {produccion} unidades', color=color, ax=ax)


ax.axhline(0, color='red', linestyle='--')
                 
ax.set_xlabel('Precio de Venta')
ax.set_ylabel('Margen de Ganancia')
ax.set_title('Margen de Ganancia por Precio de Venta y Nivel de Producción')

ax.grid(False)

plt.show()
```
